```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = sns.load_dataset("tips")
plt.figure(figsize=(8,5))

# Create scatter plot
sns.scatterplot(
    x="total_bill",
    y="tip",
    data=df,
    hue="sex",
    style="smoker",
    palette="bright"
)
#hue="sex"---- Means:Color graph based on sex column


# Add title
plt.title("Total Bill vs Tip")

# Add x-axis label
plt.xlabel("Restaurant Bill")

# Add y-axis label
plt.ylabel("Tip Amount")

# Show graph
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = sns.load_dataset("tips")

# Figure size
plt.figure(figsize=(8,5))

# Scatter plot
sns.scatterplot(
    x="total_bill",
    y="tip",
    hue="sex",
    marker="X",
    data=df
)

# Title
plt.title("Bill vs Tip")

# Labels
plt.xlabel("Total Bill")
plt.ylabel("Tip")

# Show graph
plt.show()
```

What is lineplot?

A line plot is used to show:

How something changes as another value changes

Usually used for:

Trends
Growth
Time series data
Comparisons

```python
import seaborn as sns
import matplotlib.pyplot as plt
df=sns.load_dataset("tips")
plt.figure(figsize=(8,5))

sns.lineplot(
    x="size",
    y="tip",
    color="red",
    hue="sex",
    data=df
)

plt.title("Tip Trend by Group Size")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("tips")

plt.figure(figsize=(10,6))

sns.barplot(
    x="day",
    y="total_bill",
    hue="sex",
    palette="bright",
    data=df
)

plt.title("Average Bill by Day")
plt.xlabel("Day")
plt.ylabel("Average Bill")

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("tips")

plt.figure(figsize=(10,6))

sns.countplot(
    x="day",
    hue="sex",
    palette="pastel",
    data=df
)

plt.title("Customer Count by Day")
plt.xlabel("Day")
plt.ylabel("Count")

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("tips")

plt.figure(figsize=(8,5))

sns.histplot(
    df["total_bill"]
)

plt.title("Distribution of Total Bill")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("tips")

plt.figure(figsize=(8,5))

sns.boxplot(
    x="day",
    y="total_bill",
    data=df
)

plt.title("Bill Distribution by Day")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("tips")

corr = df.corr(numeric_only=True)

plt.figure(figsize=(8,5))

sns.heatmap(
    corr,
    annot=True
)

plt.show()
```
