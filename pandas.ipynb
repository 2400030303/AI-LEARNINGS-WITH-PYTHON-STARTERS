What is Pandas?

Pandas is used for:

handling datasets
working with tables
cleaning data
analyzing data

Think of it like Excel inside Python.

```python
# A Series is a single column
#left side=index
#right side=values
import pandas as pd
s=pd.Series([10,20,30,40])
print(s)
```

```python
# A dataframe is a table.
import pandas as pd
data={
    "name":["Sai","Ram","john"],
    "marks":[90,85,95]
}
df=pd.DataFrame(data)
print(df)
print(df.columns)
print(df.shape)
# 3 rows and 2 columns
print(df.head())
print(df.tail())
print(df.tail())
print(df.info())
print(  df.describe())
```

```python
# task-1
import pandas as pd
data={
    "Name":["Sai","Ram","John","Ravi","Anu"],
    "Age":[19,20,18,21,19],
    "Marks":[90,85,95,70,88]
}
df=pd.DataFrame(data)
print(df)
print(df.shape)
print(df.columns)
print(df.info())
print(df.describe())
print(df.head())
print(df.tail())
```

```python
# selecting data
import pandas as pd
data={
    "Name":["Sai","Ram","John","Ravi","Anu"],
    "Age":[19,20,18,21,19],
    "Marks":[90,85,95,70,88]
}
df=pd.DataFrame(data)
print(df)
print(df["Name"])
print(df[["Name","Age"]])
# single columns -> single brackets and Multiple columns -. multiple brackets
print(df[df["Marks"]>85])
print(df[df["Age"]<20])
```

```python
#statistics
import pandas as pd
data={
    "Name":["Sai","Ram","John","Ravi","Anu"],
    "Age":[19,20,18,21,19],
    "Marks":[90,85,95,70,88]
}
df=pd.DataFrame(data)
print(df["Marks"].mean())
print(df["Marks"].median())
print(df["Marks"].mode())
print(df["Marks"].max())
print(df["Marks"].min())
print(df["Marks"].std())
```

**Data Cleaning**

Real datasets often contain:
* Missing values
* Duplicate rows
* Wrong Data types
* Messy values
-> Cleaning them is a huge part of AI/Data Science

```python
# isnull()-> Used to detect missing values
import pandas as pd
data={
    "Name":["Sai","Ram","John"],
    "Marks":[90,None,95]
}
df=pd.DataFrame(data)
print(df)
print(df.isnull())
# counting missing values
print(df.isnull().sum())
print(df.dropna())
print(df.fillna(98))
```

```python
# fillna() practice

import pandas as pd

data = {
    "Name": ["Sai", "Ram", "John", "Ravi"],
    "Marks": [90, None, 95, None]
}

df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)

# Fill missing values with 0
print("\nFill NaN with 0:")
print(df.fillna(0))
# Fill missing values with mean
df["Marks"] = df["Marks"].fillna(df["Marks"].mean())

print("\nFill NaN with Mean:")
print(df)

```

```python
# to check duplicate rows
import pandas as pd
data={
    "Name":["Sai","Ram","Sai","John"],
    "Marks":[90,85,90,95]
}
df=pd.DataFrame(data)
print(df.duplicated())
print(df.duplicated().sum())
print(df.drop_duplicates())
```

```python
#replace() -> used to replace values inside datasets.
import pandas as pd
data={
    "Name":["Sai","Ram","John"],
    "Result":["Pass","Fail","Pass"]
}
df=pd.DataFrame(data)
print(df)
df["Result"]=df["Result"].replace("Fail","Absent")
print(df)
```

```python
# replace() practice task
import pandas as pd
data={
    "Name":["Sai","Ram","John","Ravi"],
    "Grade":["A","B","C","B"]

}
df=pd.DataFrame(data)
print(df)
df["Grade"]=df["Grade"].replace("B","Average")
df["Grade"]=df["Grade"].replace("C","Needs Improvement")
print(df)
```

```python
# astype() -> Used to change data types.
import pandas as pd
data={
    "Age":["19","20","18"]
}
df=pd.DataFrame(data)
print(df.dtypes)
df["Age"]=df["Age"].astype(int)
print(df.dtypes)
```

```python
# sort_values -> Used to sort rows based on column values.
import pandas as pd

data = {
    "Name": ["Sai", "Ram", "John", "Ravi"],
    "Marks": [90, 85, 95, 70]
}

df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)
print(df.sort_values(by="Marks"))
print(df.sort_index())
print(df.sort_values(by="Marks",ascending=False))
print(df["Marks"].sum())
print(df["Marks"].count())
print(df["Marks"].value_counts)
```

```python
import pandas as pd
data={
    "Name":["Sai","Ram","John","Ravi"],
    "Grade":["A","B","C","B"]
}
df=pd.DataFrame(data)
print(df)
print(df["Grade"].value_counts())
```

```python
# groupby()  -> used to group similar data objects.
import pandas as pd
data={
    "Department": ["IT", "HR", "IT", "HR"],
    "Salary": [50000, 40000, 60000, 45000]
}
df=pd.DataFrame(data)
print(df.groupby("Department")["Salary"].mean())
```

```python
# drop()-> used to delete columns.
import pandas as pd
data = {
    "Name": ["Sai", "Ram", "John"],
    "Age": [19, 20, 18],
    "Marks": [90, 85, 95]
}

df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)
df=df.drop("Age",axis=1)
# axis=1 -> column operation

print(df)
df=df.rename(columns={"Marks":"Result"})
print(df)
df["Result"] = df["Result"] + 5

print(df)
df.to_csv("students.csv",index=False)
```

```python
import pandas as pd
df1 = pd.DataFrame({
    "Name": ["Sai", "Ram"]
})

df2 = pd.DataFrame({
    "Name": ["John", "Ravi"]
})

print(pd.concat([df1, df2]))
```

```python
import pandas as pd
df1 = pd.DataFrame({
    "ID": [1, 2],
    "Name": ["Sai", "Ram"]
})

df2 = pd.DataFrame({
    "ID": [1, 2],
    "Marks": [90, 85]
})

print(pd.merge(df1, df2, on="ID"))
```
