Grid

Grid helps read values clearly.

Code

plt.grid()

```python
import matplotlib.pyplot as plt
days=[1,2,3,4,5]
study=[2,4,3,5,6]
plt.plot(days,study,marker='o',linewidth=3,linestyle='-.')
plt.grid()
plt.title("Daily Study Hours")
plt.xlabel("Days")
plt.ylabel("Study")
plt.show()
#Line Width -> Makes line thicker.
#Line Style -> Changes line appearance.
#'-' → normal line;'--' → dashed;':' → dotted;'-.' → dash-dot
```

Problem

You cannot identify which line belongs to what.

So we use:

legend()

```python

import matplotlib.pyplot as plt
days=[1,2,3,4,5]
study=[2,4,3,5,6]
water=[2,3,5,4,6]
plt.plot(days,study,label='STUDY')
plt.plot(days,water,label='WATER')
plt.legend()
plt.show()
```

```python
#task
import matplotlib.pyplot as plt
days=[1,2,3,4,5]
workout=[1,2,1,2,3]
study=[3,4,5,4,6]
plt.plot(days,workout,'r',label='WORKOUT',linewidth=2,linestyle='--',marker='*')
plt.plot(days,study,'b',label='STUDY',linewidth=2,linestyle='-',marker='o')
plt.legend()
plt.grid()
plt.title("Daily Progress")
plt.xlabel("Days")
plt.ylabel("Progress")
plt.show()
```

Bar charts are used to compare categories.

Examples:
- Student marks
- Product sales
- Study hours
- App downloads

```python
import matplotlib.pyplot as plt

subjects=['Python','Java','C','AI']

hours=[5,3,2,4]

plt.figure(figsize=(8,5))

plt.bar(
    subjects,
    hours,
    color='b',
    width=0.8
)

plt.title("Study Hours Per Subject")

plt.xlabel("Subjects")

plt.ylabel("Hours")

plt.grid()

plt.show()

# Small width → thin bars; Large width → thick bars
```

```python
# Horizanatal bar charts
import matplotlib.pyplot as plt
subjects=['python','java','c','ai']
hours=[5,3,2,4]
colors=['r','b','g','orange']
plt.figure(figsize=(8,5))
plt.barh(subjects,hours,color=colors)
plt.title("Study Hours Per Subject")
plt.xlabel("Hours")
plt.ylabel("Subjects")
plt.grid()
plt.show()
# barh() creates horizontal bar chart.
```

```python
# Task1 -> Vertical bar chart
import matplotlib.pyplot as plt
apps=['Instagram','Youtube','Whatsapp','Chatgpt']
usage=[2,4,3,5]
plt.figure(figsize=(8,5))
plt.barh(apps,usage,color='g')
plt.title("Apps usage")
plt.xlabel("Usage")
plt.ylabel("Apps")
plt.grid()
plt.show()
```

```python
# first pie chart
import matplotlib.pyplot as plt
subjects=['python','java','c','ai']
hours=[5,3,2,4]
plt.pie(hours,labels=subjects,autopct='%1.1f%%')
plt.show()
```

autopct displays percentages.

%1.1f%%
→ shows percentage with 1 decimal place

```python
# Task
import matplotlib.pyplot as plt
apps=['Instagram','YouTube','WhatsApp','ChatGPT']

usage=[2,4,3,5]
plt.figure(figsize=(8,5))
plt.pie(usage,labels=apps,autopct='%1.1f%%')
plt.title("Apps Usage")
plt.show()
```

```python
# First Scatter plot
import matplotlib.pyplot as plt
study=[1,2,3,4,5]
marks=[40,50,60,70,80]
plt.scatter(study,marks,color='red',marker='*')
plt.title("Study vs Marks")
plt.xlabel("Study")
plt.ylabel("Marks")
plt.grid()
plt.show()
```

scatter() creates separate points.

Unlike plot(), points are not connected.

```python
# Histogram
import matplotlib.pyplot as plt


ages=[18,19,20,21,22,23,24,25,18,19,20,21,22,23]
plt.figure(figsize=(8,5))

plt.hist(
    ages,
    bins=5,
    color='orange'
)

plt.title("Ages Distribution")

plt.xlabel("ages")

plt.ylabel("Frequency")

plt.grid()

plt.show()
```

Histogram groups data into ranges.

It shows frequency distribution.

```python
# Using Pandas Dataset
import pandas as pd
import matplotlib.pyplot as plt
data={
    'Days':[1,2,3,4,5],
    'Study':[2,4,3,5,6]
}
df=pd.DataFrame(data)
print(df)
plt.plot(df['Days'],df['Study'],marker='*',color='r',linewidth=2,linestyle='--')
plt.title("Daily Study Hours")
plt.xlabel("Days")
plt.ylabel("Study")
plt.grid()
plt.savefig("graph.png")
plt.show()
#savefig() saves the graph as an image file.
```
